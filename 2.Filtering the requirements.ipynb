```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
#Paths
raw_path = "datasets/raw"
processed_path = "datasets/processed"
presentation_path = "datasets/presentation"
```

### Creating dataframes of cleaned data

```python
features_df = pd.read_csv(f"{processed_path}/processed_features.csv")
tracks_df = pd.read_csv(f"{processed_path}/processed_tracks.csv")
artists_df = pd.read_csv(f"{processed_path}/processed_artitst.csv")
```

### 1.Top Artists by Popularity:


```python
artists_df.head()
```

```python
top_artists_df = artists_df.sort_values(by = 'popularity', ascending = False).head(10)
top_artists_df
```

```python
top_artists_df.to_csv(f"{presentation_path}/top_artists_by_popularity.csv", index =False)
```

### 2.Top Tracks by Audio Features: 

```python
tracks_df.head()
```

```python
top_tracks_df = tracks_df.sort_values(by = ['danceability','energy','acousticness'], ascending= [False, False, True]).head(10)
top_tracks_df
```

```python
top_tracks_df.to_csv(f"{presentation_path}/top_tracks_by_features.csv", index =False)
```

### 3.Top Tracks by Popularity: 

```python
top_tracks_by_popularity = tracks_df.sort_values(by = ['popularity'], ascending=False).head(10)
```

```python
top_tracks_by_popularity.to_csv(f"{presentation_path}/top_tracks_by_popularity.csv", index =False)
```

```python
top_tracks_by_popularity
```

### 4. Count the songs under the Specific genres

```python
features_df
```

```python
genre_group_by = features_df.groupby('genre').agg(count =('track_id','count')).reset_index()
genre_group_by = genre_group_by.sort_values(by=['count'], ascending=False)
```

```python
genre_group_by.to_csv(f"{presentation_path}/genre_group_by_count.csv", index =False)
```

```python
genre_group_by
```

### 5. Analysis of Releases of tracks by year:

```python
print(tracks_df['release_date'].dtypes)
```

```python
tracks_df.head()
```

```python
new_tracks_df = tracks_df
```

```python
new_tracks_df['year'] = pd.DatetimeIndex(new_tracks_df['release_date']).year
```

```python
new_tracks_df.head(2)
```

```python
yearly_tracks_released_df = new_tracks_df.groupby('year').agg(count = ('id','count')).reset_index().sort_values(by =['count'], ascending=False)
```

```python
yearly_tracks_released_df.to_csv(f"{presentation_path}/yearly_tracks_released_count.csv", index =False)
yearly_tracks_released_df
```

### 6.Track Duration Distribution:


```python
avg_dur_track = new_tracks_df.groupby('year').agg( mean_duration = ('duration_ms','mean')).reset_index()
avg_dur_track['mean_minutes'] = (avg_dur_track['mean_duration']/1000)/60
avg_dur_track = avg_dur_track.sort_values(by = ['year'], ascending=False)
```

```python
avg_dur_track.to_csv(f"{presentation_path}/avg_dur_track.csv", index =False)
avg_dur_track
```
