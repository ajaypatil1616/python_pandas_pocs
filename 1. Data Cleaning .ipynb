```python
tracks_df['release_date'] = tracks_df['release_date'].astype("str")
```

```python
tracks_df['release_date']
```

```python
def clean_release_date(date):
    parts = date.split('-')
    if len(parts) == 1:
        return f"{parts[0]}-01-01"
    elif len(parts) ==2:
        return f"{parts[0]}-{parts[1]}-01"
    else:
         return date
```

```python
tracks_df['release_date'] = tracks_df['release_date'].apply(clean_release_date)
```
